```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
  f1 = 3/(1+(x-2)**2+4*(y-3)**2)
  f2 = 4/(1+x**2+y**2)
  return f1 - f2

x = np.linspace(-2,4)
y = np.linspace(-1,4)
(X,Y) = np.meshgrid(x,y)
Z = f(X,Y)
plt.set_cmap('rainbow')
plt.contour(X, Y, Z, linewidths=3)
plt.colorbar()
plt.grid('True')
```
