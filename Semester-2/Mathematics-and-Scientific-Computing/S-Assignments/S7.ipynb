```python
import numpy as np
import matplotlib.pyplot as plt

(√°r, hiti, √∫rk) = np.loadtxt("https://cs.hi.is/python/hiti-urkoma.txt").T

(a, b) = np.polyfit(hiti, √∫rk, deg=1)

xp = np.linspace(min(hiti), max(hiti))
yp = a*xp + b

plt.figure(figsize=(12,6))
plt.scatter(hiti, √∫rk, label='√Års√∫rkoma vs Hiti')
plt.plot(xp, yp, c='r', label='L√≠na bestu j√∂fnu')
plt.xlabel('Hiti (¬∞C)')
plt.ylabel('√Års√∫rkomu (mm)')
plt.title('√Års√∫rkoma mi√∞a√∞ vi√∞ hita')
plt.legend()
plt.show()
```

# 29 Fjarl√¶g√∞ milli borga

```python
from math import sin, cos
import math

def Matrix(R, l, b):
    a = R * sin(l) * cos(b)
    b = R * cos(l) * cos(b)
    c = R * sin(b)
    return a,b,c

R = 6370
ReyLen = np.deg2rad(22)
ReyWidth = np.deg2rad(65)

LonLen = np.deg2rad(0)
LonWidth = np.deg2rad(51.5)

a, b, c = Matrix(R, ReyWidth, ReyLen)
d, e, f = Matrix(R, LonWidth, LonLen)

ReyMatrix = np.array([a, b, c])
LonMatrix = np.array([d, e, f])

Calculations = np.rad2deg(math.acos((ReyMatrix @ LonMatrix)/((la.norm(ReyMatrix))*(la.norm(LonMatrix)))))
Distance = np.deg2rad(Calculations)*R

print(f"Horni√∞ milli Reykjav√≠kur og London m√¶lt fr√° mi√∞ju jar√∞ar er {round(Calculations,3)} gr√°√∞ur")
print(f"Fjarl√¶g√∞in milli Reykjav√≠kur og London er {round(Distance,2)} km")

```
