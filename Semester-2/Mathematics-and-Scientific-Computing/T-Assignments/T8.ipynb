```python
import numpy as np
from math import nan, sin, cos
from matplotlib import pyplot as plt

def teikna_fylki(A):
   # Teiknar 2 x n flatarmyndarfylki
   plt.plot(A[0],A[1], lw=3, color='k')

def hlira(A, h):
   # Leggur h vi alla dlka A
   return A + np.reshape(h,(2,1))

M = np.array([[0,3,3,0,0,nan,1,2], [0,0,2,2,0,nan,1,1]])

teikna_fylki(M)
plt.show()

def skekking(k, a):
   skekkingin1 = np.array([[1,sin(k)], [0, 1]])
   skkkingin2 = np.array([[1,0],[sin(k), 1]])
   return skekkingin1@a
a = skekking(30, M)

teikna_fylki(a)
plt.show()

def snuningsfylki(c,v):
   snuningur = np.array([[cos(c), -sin(c)], [sin(c),cos(c)]])
   return snuningur@v

b = snuningsfylki(360, M)
teikna_fylki(b)
plt.show()

stortfylki = np.array([[0,5,5,0,0], [0,0,3,3,0]])
litidfylki = np.array([[1,2,2,1,1], [1,1,2,2,1]])


teikna_fylki(stortfylki)
teikna_fylki(litidfylki)
plt.show()

thrihyrningur = np.array([[1,3,4,1],[0,2,0,0]])
teikna_fylki(thrihyrningur)
plt.show()
```
