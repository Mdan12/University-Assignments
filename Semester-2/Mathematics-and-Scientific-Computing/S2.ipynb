Simpsons-formÃºlan er eftirfarandi:
$$
\ \int{a}^{b} f(x) \, dx \approx \frac{\Delta x}{3}\left(f(x_0) + 4f(x_1)+2f(x_2)+4f(x_3)+2f(x_4)+\cdots+4f(x_{n-1}) + f(x_{n})\right)

SkrifiÃ° fall simpson(f,a,b,n) sem nÃ¡lgar heildiÃ° af f frÃ¡ a til b meÃ° samsettri Simpsons-reglu meÃ° n hlutbilum.

PrÃ³fiÃ° meÃ° heildunum (*) og (**) Ãºr A-liÃ° meÃ° 4 hlutbilum. Ef rÃ©tt er forritaÃ° Ã¦tti aÃ° fÃ¡st (âˆ—)

````python
from math import sin
from math import cos

def simpson(f,a,b,n):
    dx = (b-a)/n
    s = f(a)+f(b)
    for i in range (1,n):
        xi = a+i*dx
        if i%2==0:
            s += 2*f(xi)
        else:
            s += 4*f(xi)
    return (dx/3)*s  

from math import sin, e

def g(x): return sin(x)/x
def h(x): return e**x
def z(x): return sin(2*x)*cos(2*x)

I1 = simpson(g,1,2,2)
I2 = simpson(h,0,1,4)
I3 = simpson(z,24,93,11)
print(f"Simpsons-reglan Ã¡ fyrri jÃ¶fnu tveimur hlutblium hefur svariÃ° {I1}, hinsvegar hefur seinni jafnan svariÃ° {I2} ef skipt er Ã­ 4 hlutbita.")
print(f"Ã‰g er fÃ¦ddur 24.11.1993 og nota Ã¾vÃ­ jÃ¶fununa meÃ° Ã¾eim dagsetningum. SvariÃ° er Ã¾vÃ­ {I3}")
````

# BÃ³lurÃ¶Ã°un
HÃ©r er reiknirit sem raÃ°ar n-staka lista x meÃ° bÃ³lurÃ¶Ã°un bubble sort, sem snÃ½st um aÃ° rÃºlla Ã­ gegn um stÃ¶kin og ef tvÃ¶ samliggjandi stÃ¶k eru Ã­ Ã¶fugri rÃ¶Ã° Ã¾Ã¡ er vÃ­xlaÃ° Ã¡ Ã¾eim. Ãetta er endurtekiÃ° Ã¾ar til Ã¶ll stÃ¶kin eru Ã­ rÃ¶Ã°. Minnstu stÃ¶kin bobbla smÃ¡m saman eins og loftbÃ³lur fremst Ã­ listann.
```
vÃ­xlaÃ° = satt
meÃ°an vÃ­xlaÃ°
   vÃ­xlaÃ° = Ã³satt
   fyrir i=1,...,n-1:
      ef x[i-1] > x[i] Ã¾Ã¡
         vÃ­xla Ã¡ x[i-1] og x[i]
         vÃ­xlaÃ° = satt
```

ÃÃ½Ã°iÃ° Ã¾etta reiknirit yfir Ã­ Python-fall bÃ³lurÃ¶Ã°un(x). AthugiÃ° aÃ° til aÃ° vÃ­xla Ã¡ breytum x og y mÃ¡ nota (x,y) = (y,x). PrÃ³fiÃ° meÃ° Ã¾vÃ­ aÃ° raÃ°a listanum [3,8,1,2,5,4].


````python
def bolurodun(x):
    reversed = True
    n = len(x)
    while (reversed):
        reversed = False
        for i in range (1,n):
            if x[i-1]> x[i]:
                x[i-1], x[i] = x[i], x[i-1]
                reversed = True
    return x

x = [3,8,1,2,5,4]

print(f"Ef viÃ° notum bÃ³lurÃ¶Ã°un Ã¡ eftirfarandi lista: {x} Ã¾Ã¡ fÃ¡um viÃ° Ãºtkommuna {bolurodun(x)}")
````
